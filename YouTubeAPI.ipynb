```python
# Adele's channel
channel_id = 'UCsRM0YB_dabtEPGPTKo-gcw'
```

```python
# GET the channel statistics 

url = 'https://www.googleapis.com/youtube/v3/channels?part=statistics&id='+channel_id+'&key='+my_key
```

```python
import urllib.request
contents = urllib.request.urlopen(url).read()
```

```python
print(contents.decode("utf-8") )
```

```python
# parse json
import json
channel = json.loads(contents.decode("utf-8"))
channel
```

```python
# GET comments threads of a channel

url = 'https://www.googleapis.com/youtube/v3/commentThreads?part=snippet&allThreadsRelatedToChannelId='+channel_id+'&key='+my_key
contents = urllib.request.urlopen(url).read()
print(contents.decode("utf-8"))
```

```python
comments = json.loads(contents.decode("utf-8"))
```

```python
print(str(len(comments['items']))+' comment threads are extracted for this channel')
```

```python
print('Comment time:')
for i in range(len(comments['items'])):
    print(comments['items'][i]['snippet']['topLevelComment']['snippet']['publishedAt'])
```

```python
# GET comments for specific videos
# VideoID: hLQl3WQQoQ0 // https://www.youtube.com/watch?v=hLQl3WQQoQ0 // Adele - Someone Like You

url = "https://www.googleapis.com/youtube/v3/commentThreads?key="+my_key+"&textFormat=plainText&part=snippet&videoId=hLQl3WQQoQ0&maxResults=20"
contents = urllib.request.urlopen(url).read()
print(contents.decode("utf-8"))
```

```python
comments = json.loads(contents.decode("utf-8"))
```

```python
print('Comment time:')
for i in range(len(comments['items'])):
    print(comments['items'][i]['snippet']['topLevelComment']['snippet']['publishedAt'])
```

```python
# To Get sequential results, use &pageToken={nextPageToken}
```

```python
url = "https://www.googleapis.com/youtube/v3/commentThreads?key="+my_key+"&textFormat=plainText&part=snippet&videoId=hLQl3WQQoQ0"+"&pageToken="+comments['nextPageToken']
contents = urllib.request.urlopen(url).read()
print(contents.decode("utf-8"))
```

```python
comments = json.loads(contents.decode("utf-8"))
print('Comment time:')
for i in range(len(comments['items'])):
    print(comments['items'][i]['snippet']['topLevelComment']['snippet']['publishedAt'])
```
