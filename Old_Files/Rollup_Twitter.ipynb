```python
import pandas as pd
```

```python
time_range = pd.date_range(start='1/1/2015', end='6/30/2019', freq='d').to_frame(index=False)
```

```python
time_range.rename(columns={0: 'Time Period'}, inplace=True)
```

```python
master_artists_list = pd.read_csv('master_artists_list.csv', usecols=['twitter'])
```

```python
time_range['join_key'] = 0
master_artists_list['join_key'] = 0
```

```python
artists_time_range = master_artists_list.merge(time_range, how='left', on = 'join_key')
artists_time_range.drop('join_key',1, inplace=True)
artists_time_range.rename(columns={0: 'Time Period'}, inplace=True)
```

```python
twitter_file = pd.read_csv('twitter_features.csv', usecols=['Twitter_Handle', 'created', 'crowdtangle_score'])
```

```python
twitter_file['created'] = pd.to_datetime(twitter_file['created'])
twitter_file['Created_Day'] = twitter_file.created.dt.floor('d')
twitter_file.head()
```

```python
df_aggregate = twitter_file.groupby(["Twitter_Handle", 'Created_Day'], as_index=False).agg({"created": "count"})
df_aggregate.rename(columns={'created': 'Posts In Day'}, inplace=True)
df_aggregate.head()
```

```python
df_merged = artists_time_range.merge(df_aggregate, how='left', left_on=['twitter','Time Period'], right_on=['Twitter_Handle','Created_Day'])
df_merged[df_merged['Twitter_Handle'].notnull()]
```

```python
df_merged['Posts In Day'] = df_merged['Posts In Day'].fillna(0)
```

```python
df_merged.head()
```

```python
df_merged['last_7_days'] = df_merged[['Posts In Day']].groupby(df_merged['twitter']).apply(lambda g: g.rolling(7, min_periods=1).sum())
df_merged['last_30_days'] = df_merged[['Posts In Day']].groupby(df_merged['twitter']).apply(lambda g: g.rolling(30, min_periods=1).sum())
```
