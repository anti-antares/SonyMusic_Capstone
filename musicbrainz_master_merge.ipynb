```python
from datetime import datetime

def days_between(d1, d2):
    d1 = datetime.strptime(d1, "%Y-%m-%d")
    d2 = datetime.strptime(d2, "%Y-%m-%d")
    return abs((d2 - d1).days)
```

```python
def find_min_diff(artist_name,created):
#     print(artist_name,created)
    lst = artist_dict[artist_name]
    min_days = min([days_between(x,created.strftime('%Y-%m-%d')) for x in lst])
    return min_days
```

```python
t1 = time.time()
df_temp['num_days_to_track_release'] = df_temp.apply(lambda x: find_min_diff(x['artist_name_y'],x['created']),axis=1)
print("time taken",(time.time() - t1))
```
