```python
import pandas as pd
import numpy as np
from datetime import timedelta
```

```python
ydata = pd.read_csv('youtube_data.csv').iloc[:,1:]
```

```python
ydata.head()
```

```python
ydata = ydata.sort_values(by=['artist','time'])
```

```python
ydata['time'] = pd.to_datetime(ydata['time'])
type(ydata.iloc[1,2])
```

```python
ydata.iloc[1,2]-ydata.iloc[0,2]
```

```python
ydata['since_last_post'] = ydata.groupby('artist')['time'].diff()
```

```python
ydata.head()
```

```python
def get_hour(d):
    return d.hour
ydata["hour"] = ydata["time"].apply(get_hour)
```

```python
ydata['hour'].unique()
```

```python
def time_of_day_creation(x):
    '''
    Takes in an integer and returns one of four time slots:
    ['12am-6am', '6am-12pm', '12pm-6pm', '6pm-12am']
    '''
    
    slots_out = ['12am-6am', '6am-12pm', '12pm-6pm', '6pm-12am']
    slot0 = np.arange(0,6)
    slot1 = np.arange(6,12)
    slot2 = np.arange(12,18)
    slot3 = np.arange(18,24)

    if x in slot0:
        return slots_out[0]
    elif x in slot1:
        return slots_out[1]
    elif x in slot2:
        return slots_out[2]
    elif x in slot3:
        return slots_out[3]
    else:
        return 'NaN'
```

```python
ydata['time_of_day']=ydata['hour'].apply(time_of_day_creation)
```

```python
ydata['within_week_release'] = 0
ydata['within_month_release'] = 0
```

```python
ydata = pd.concat([ydata ,pd.get_dummies(ydata ['time_of_day'],prefix='tod')],axis=1)
```

```python
ydata
```

```python
ydata['post_last_week'] = 0
ydata['post_last_month'] = 0
```

```python
ydata=ydata.reset_index(drop=True)
ydata.head()
```

```python
timedelta(days=30)
```

```python
for i in range(1,len(ydata)):
    if (i+1)%200==0:
        print(str(i+1)+ ' has been completed!')
    posts=0
    for j in range(i-1,-1,-1):
        if ydata['artist'][i]!=ydata['artist'][j]:
            break
        if ydata['time'][i]-ydata['time'][j]>timedelta(weeks=1):
            break
        else:
            posts+=1
    ydata['post_last_week'][i]=posts
```

```python
for i in range(1,len(ydata)):
    if (i+1)%200==0:
        print(str(i+1)+ ' has been completed!')
    posts=0
    for j in range(i-1,-1,-1):
        if ydata['artist'][i]!=ydata['artist'][j]:
            break
        if ydata['time'][i]-ydata['time'][j]>timedelta(days=30):
            break
        else:
            posts+=1
    ydata['post_last_month'][i]=posts
```

```python
ydata.to_csv("youtube_data_custom.csv")
```
