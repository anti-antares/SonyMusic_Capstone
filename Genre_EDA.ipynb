```python
import pandas as pd
import numpy as np
```

```python
a_data = pd.read_csv('Combined_Social_Data_version2.csv')
```

```python
a_data.tail()
```

```python
type(a_data.time_since_last_post[0])
```

```python
a_data.groupby(['artist_name_y','Facebook','Instagram','Twitter','Youtube']).mean()
```

```python
facebook_data = a_data[a_data.Facebook==1].groupby(['artist_name_y']).mean()
facebook_data.head()
```

```python
facebook_data.columns
```

```python
facebook_data = facebook_data.filter(items = ['age','years_active','count_of_followers','time_since_last_post',
                                              'social_engagement_score','created_12am-6am','created_12pm-6pm',
                                             'created_6am-12pm','created_6pm-12am','crowdtangle_score','description_length',
                                             'time_since_last_post','type_link','type_photo','type_text','type_video','genre_blues',
                                              'genre_country','genre_electronic','genre_folk','genre_hip_hop','genre_jazz','genre_pop'
                                              ,'genre_r_and_b','genre_rock','genre_spiritual','genre_world_music'
                                             ])
facebook_data.head()
```

```python
facebook_data['count_scale']=facebook_data['count_of_followers'].apply(np.log10)
import matplotlib as plt
facebook_data.boxplot(column=['age', 'years_active'],by='genre_hip_hop')
```

```python
facebook_data.boxplot(column=['count_scale'],by='genre_hip_hop')
```

```python
facebook_data.boxplot(column=['count_scale'],by='genre_hip_hop')
```

```python
facebook_data[facebook_data['social_engagement_score']<0.05].boxplot(column=['social_engagement_score'],by='genre_hip_hop')
```

```python
facebook_data.boxplot(column=['description_length'],by='genre_hip_hop')
```

```python
facebook_data[facebook_data['social_engagement_score']<0.05].plot.scatter(x='description_length',y='social_engagement_score',c='genre_hip_hop',colormap='viridis')
```
